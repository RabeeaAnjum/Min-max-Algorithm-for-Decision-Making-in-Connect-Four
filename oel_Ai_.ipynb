```python
import math

# Define the Connect Four board
ROWS = 6
COLS = 7
board = [[" " for _ in range(COLS)] for _ in range(ROWS)]  # Empty board

# Helper function to print the board
def print_board(board):
    for row in board:
        print("|".join(row))
        print("-" * 13)

# Check if a move results in a win
def check_winner(board):
    # Horizontal, Vertical, Diagonal (both directions)
    for row in range(ROWS):
        for col in range(COLS):
            if board[row][col] == " ":
                continue
            player = board[row][col]
            # Horizontal check
            if col + 3 < COLS and all(board[row][col + i] == player for i in range(4)):
                return player
            # Vertical check
            if row + 3 < ROWS and all(board[row + i][col] == player for i in range(4)):
                return player
            # Diagonal check (down-right)
            if row + 3 < ROWS and col + 3 < COLS and all(board[row + i][col + i] == player for i in range(4)):
                return player
            # Diagonal check (up-right)
            if row - 3 >= 0 and col + 3 < COLS and all(board[row - i][col + i] == player for i in range(4)):
                return player
    # Check for draw (board full)
    if all(board[row][col] != " " for row in range(ROWS) for col in range(COLS)):
        return "Draw"
    return None  # No winner yet

# Minimax algorithm
def minimax(board, depth, is_maximizing):
    winner = check_winner(board)
    if winner == "X":
        return 1  # Maximizing player (computer) wins
    elif winner == "O":
        return -1  # Minimizing player (human) wins
    elif winner == "Draw":
        return 0  # It's a draw

    if is_maximizing:
        best_score = -math.inf
        for col in range(COLS):
            for row in range(ROWS - 1, -1, -1):
                if board[row][col] == " ":
                    board[row][col] = "X"  # Try the move for computer
                    score = minimax(board, depth + 1, False)
                    board[row][col] = " "  # Undo the move
                    best_score = max(score, best_score)
                    break
        return best_score
    else:
        best_score = math.inf
        for col in range(COLS):
            for row in range(ROWS - 1, -1, -1):
                if board[row][col] == " ":
                    board[row][col] = "O"  # Try the move for human
                    score = minimax(board, depth + 1, True)
                    board[row][col] = " "  # Undo the move
                    best_score = min(score, best_score)
                    break
        return best_score

# Find the best move for the computer
def best_move(board):
    best_score = -math.inf
    move = None
    for col in range(COLS):
        for row in range(ROWS - 1, -1, -1):
            if board[row][col] == " ":
                board[row][col] = "X"  # Try the move for computer
                score = minimax(board, 0, False)
                board[row][col] = " "  # Undo the move
                if score > best_score:
                    best_score = score
                    move = col
                break
    return move

# Play the game
def play_game():
    print("Welcome to Connect Four!")
    print_board(board)

    while True:
        # Human move (player O)
        player_move = int(input("Enter your move (1-7): ")) - 1
        if player_move < 0 or player_move >= COLS or board[0][player_move] != " ":
            print("Invalid move. Try again.")
            continue
        # Find the lowest empty row in the column
        for row in range(ROWS - 1, -1, -1):
            if board[row][player_move] == " ":
                board[row][player_move] = "O"
                break
        print_board(board)

        # Check for winner or draw
        winner = check_winner(board)
        if winner:
            if winner == "O":
                print("You win!")
            elif winner == "X":
                print("Computer wins!")
            else:
                print("It's a draw!")
            break

        # Computer move (player X)
        print("Computer is making a move...")
        move = best_move(board)
        # Find the lowest empty row in the column
        for row in range(ROWS - 1, -1, -1):
            if board[row][move] == " ":
                board[row][move] = "X"
                break
        print_board(board)

        # Check for winner or draw
        winner = check_winner(board)
        if winner:
            if winner == "O":
                print("You win!")
            elif winner == "X":
                print("Computer wins!")
            else:
                print("It's a draw!")
            break

# Run the game
play_game()
```
